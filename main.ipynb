```python
from datetime import date

import pandas as pd
from IPython.display import Markdown

from data import PROGRAMS
from models import AVAILABILITY_TYPES, COUNTRIES, Semester
```

```python
display(Markdown(f"## This notebook is run on {date.today()}"))
```

```python
TARGET_SEMESTER = Semester(starting_year="2026", season="winter")
```

```python
programs = pd.DataFrame(
    [
        {
            "name": p.name,
            "university": p.university_name,
            "country": p.country,
            "availability": p.get_availability(TARGET_SEMESTER),
            "application_fee": p.application_fee,
            "window_start": p.get_days_until_window_starts(TARGET_SEMESTER),
            "window_end": p.get_days_until_window_ends(TARGET_SEMESTER),
            "specializations": p.specializations,
        }
        for p in PROGRAMS
    ]
)
```

```python
for a in AVAILABILITY_TYPES:
    display(Markdown(f"## {a.upper()}"))
    display(programs[programs["availability"] == a])
```

```python
display(Markdown("## Applicable Programs Sorted by Days until Window Closes"))

for country in COUNTRIES:
    display(Markdown(f"### {country}"))
    with pd.option_context("display.max_colwidth", None):
        display(
            programs[(programs["window_end"] >= 0) & (programs["country"] == country)]
            .assign(
                specializations=lambda d: d["specializations"].map(
                    lambda x: ", ".join(x) if isinstance(x, (list, tuple, set)) else x
                )
            )
            .sort_values(by=["window_end", "application_fee", "window_start"])
        )

# with pd.option_context("display.max_colwidth", None):
#     display(Markdown("## Applicable Programs' Specialiaztions"))
#     display(
#         programs[programs["window_end"] >= 0].assign(
#             specializations=lambda d: d["specializations"].map(
#                 lambda x: ", ".join(x) if isinstance(x, (list, tuple, set)) else x
#             )
#         )[["name", "university", "specializations"]]
#     )
```
